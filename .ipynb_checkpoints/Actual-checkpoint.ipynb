```python
# Create VideoCapture object
cap = cv2.VideoCapture(1)  # 0 for default webcam, or specify the index of the webcam
cap.set(3,1280)
cap.set(4,720)
while True:
    # Read frame from webcam
    ret, frame = cap.read()
    results=model(frame,stream=True)
    for r in results:
        boxes=r.boxes
        for box in boxes:
            # for bounding box
            x1,y1,x2,y2=box.xyxy[0]
            x1,y1,x2,y2=int(x1),int(y1),int(x2),int(y2)
            # cv2.rectangle(frame,(x1,y1),(x2,y2),(255,0,255),3)
            w,h=x2-x1,y2-y1
            cvzone.cornerRect(frame,(x1,y1,w,h),colorR=(255, 0, 255),
    colorC=(0, 255, 0))
            # for confidence
            conf=math.ceil((box.conf[0]*100))/100
            print(conf)
            # class names
            cls=int(box.cls[0]) # since the value was floating we have converted into integer value
            cvzone.putTextRect(frame,f'{coco_classes[cls]} {conf}',(max(0,x1),max(35,y1)))
            # cvzone.putTextRect(frame,f'{conf}',(max(0,x1),max(35,y1)))

            
    
    # Display the frame
    cv2.imshow('Webcam', frame)
    
    # Wait for key press (display frame for 1 millisecond)
    if cv2.waitKey(1) & 0xFF == ord('q'):  # Press 'q' to exit
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
